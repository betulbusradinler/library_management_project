```python
class Validation:
      def int_type_casting(value):
        try:
          is_int = int(value)
          return is_int
        except ValueError:
            print("Please enter a NUMERIC value.\n")


      def check_value_range(value):
        if value<1 :
          return False
        elif value>4 :
          return False
        else :
          return True

class Library:
      def __init__(self) :
        try:
          self.file = open("books.txt", "a+")
          if self.file.tell() == 0:
                books_colums = "book_name,author,release_date,number_of_pages\n"
                self.file.write( books_colums)
        except FileNotFoundError:
            print("File was not found")

      def initialize_file(self):
            if self.file.tell() == 0:
                books_columns = "book_name,author,release_date,number_of_pages\n"
                self.file.write(books_columns)

      def __del__(self):
        self.file.close()

      def list_books(self):
            self.file.seek(0)  # Go to the beginning of the file
            lines = self.file.readlines()[1:]  # Skip the first line (column headers)
            for line in lines:
                book_info = line.strip().split(',')
                book_name = book_info[0]
                author = book_info[1]
                print(f"Book: {book_name}, Author: {author}")

      def add_book(self) :
          try:
            title = input("Enter the title of the book: ")
            author = input("Enter the author of the book: ")

            while True:
              release_year = input("Enter the release year of the book: ")
              is_int = Validation.int_type_casting(release_year)
              if type(is_int) is int :
                break
              else :
                 continue

            while True:
                pages = input("Enter the number of pages of the book: ")
                is_int = Validation.int_type_casting(pages)
                if type(is_int) is int :
                  break
                else :
                  continue

            new_add_book = "{}, {}, {}, {}".format(title, author,release_year,pages)
            self.file.write(new_add_book)
            self.file.flush() # bu kodu kullanmadığım da son eklediğim eleman tam olarak eklenmiyordu.
            print("Book added successfully.")

          except FileNotFoundError :
            print("File was not found")

      def remove_book(self):
                      book_to_remove = input("Enter the name of the book to remove: ")
                      lines = self.file.readlines()
                      self.file.seek(0)
                      books = [line.strip() for line in lines[1:]]  # Skip the column header
                      books_to_keep = []
                      book_removed = False
                      for book in books:
                          if book_to_remove not in book:
                              books_to_keep.append(book)
                          else:
                              book_removed = True

                      if not book_removed:
                          print(f"Book '{book_to_remove}' not found in library.")
                      else:
                          with open("books.txt", "w") as file:
                              file.write("book_name,author,release_date,number_of_pages\n")  # Rewrite the column header
                              for book in books_to_keep:
                                  file.write(book + '\n')
                          print(f"Book '{book_to_remove}' removed successfully.")


lib = Library()
lib.initialize_file()

while True:
        print( "***MENU***\n1) List Books\n2) Add Book\n3) Remove Book\n4) Quit")
        choice = input("Enter your choice: ")
        try:
            type_casting = Validation.int_type_casting(choice)

            if type(type_casting) is int  :

               range_value = Validation.check_value_range(type_casting)

               if range_value == False :
                  print("Please enter a value between 1 and 4 \n");

               else :
                  if type_casting == 1 :
                      lib.list_books()

                  elif type_casting == 2 :
                       lib.add_book()

                  elif type_casting == 3 :
                       lib.remove_book()

                  elif type_casting == 4:
                      del lib
                      break
                  else:
                      print("Invalid choice. Please enter a valid option.")

        except ValueError :
            print("Invalid choice. Please enter a valid option.")
```
